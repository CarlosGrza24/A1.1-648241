Para elaborar la gráfica se utilizó la librería *plt* para graficar. La función *plt.scatter* grafica los datos que seleccionas y la función *plt.figure* sirve para definir el tamaño de la gráfica que vamos hacer.


```python
import matplotlib.pyplot as plt
y= datos.NivelDeObesidad[datos.Sexo == "Male"]
x= datos.Peso[datos.Sexo == "Male"]

plt.scatter(x = datos.Peso , y = datos.NivelDeObesidad , alpha = 0.05)
plt.show()
```
